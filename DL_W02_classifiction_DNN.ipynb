#Classifiction DNN 分類深度神經網路

**什麼是分類深度神經網路？**

分類深度神經網路是一種使用深度學習技術來將輸入資料分配到預定義類別的神經網路。與迴歸任務預測連續數值不同，分類任務的目標是預測輸入資料屬於哪個類別。例如，判斷一張圖片是貓還是狗，或者判斷一封電子郵件是否為垃圾郵件。

**分類深度神經網路如何運作？**

分類深度神經網路的架構與迴歸深度神經網路類似，但輸出層和損失函數會根據分類任務的特性進行調整。主要組成部分包括：

1.  輸入層（Input Layer）： 接收輸入特徵。
2.  隱藏層（Hidden Layers）： 多個相互連接的神經元層，用於學習資料中的複雜模式。
3.  激活函數（Activation Functions）： 應用於隱藏層的神經元，引入非線性。常見的激活函數包括 ReLU、Sigmoid 和 Tanh。
4.  輸出層（Output Layer）： 產生每個類別的預測概率或類別標籤。
5.  二元分類（Binary Classification）： 輸出層通常只有一個神經元，使用 Sigmoid 激活函數，輸出一個介於 0 和 1 之間的概率值，表示輸入屬於正類別的可能性。
6.  多類別分類（Multiclass Classification）： 輸出層包含與類別數量相同的神經元，通常使用 Softmax 激活函數。Softmax 將每個神經元的輸出轉換為一個介於 0 和 1 之間的概率值，且所有類別的概率之和為 1。每個概率值表示輸入屬於對應類別的可能性。
7.  損失函數（Loss Function）： 用於衡量模型預測的類別與實際類別之間的差異。在分類任務中，**常用的損失函數包括：**
  * 二元交叉熵損失（Binary Cross-Entropy Loss）： 用於二元分類。
  * 多類別交叉熵損失（Categorical Cross-Entropy Loss）： 用於多類別分類。

**訓練過程：**

分類深度神經網路的訓練過程與迴歸深度神經網路類似，包括前向傳播、計算損失、反向傳播和優化。目標是最小化損失函數，使模型的預測結果與實際標籤盡可能一致。

**迴歸深度神經網路的優點：**
1.  強大的特徵學習能力： 能夠自動學習資料中用於區分不同類別的複雜特徵。
2.  處理高維度資料： 擅長處理具有大量輸入特徵的分類問題。
3.  捕捉非線性決策邊界： 透過多個隱藏層和非線性激活函數，可以學習複雜的非線性.決策邊界，從而有效地區分不同的類別。
4.  適用於複雜的分類問題： 在圖像識別、自然語言處理、語音識別等複雜的分類任務中表現優異。
**迴歸深度神經網路的缺點：**
1.  需要大量的標註資料： 訓練一個高性能的分類深度神經網路通常需要大量的帶有正確標籤的資料。
2.  訓練成本高： 模型結構複雜，訓練時間可能較長，需要較高的計算資源。
3.  容易過擬合： 在訓練資料不足或模型過於複雜的情況下，容易發生過擬合。需要使用正則化技術（例如 L1、L2 正則化、Dropout）和資料增強等方法來緩解。
4.  模型解釋性較差： 深度神經網路的決策過程通常難以解釋。
5.  類別不平衡問題： 當不同類別的資料量差異很大時，模型可能傾向於預測數量較多的類別。需要使用特殊的技巧來處理類別不平衡問題，例如調整類別權重、過採樣或欠採樣。
6.  超參數調整敏感： 模型的性能對超參數的選擇非常敏感，需要仔細調整。
**分類深度神經網路的常見應用：**

1.  圖像識別： 例如，識別圖片中的物體（貓、狗、汽車等）、人臉識別。
2.  自然語言處理： 例如，文本分類（情感分析、主題分類、垃圾郵件檢測）、命名實體識別。
3.  語音識別： 將語音轉換為文字，並對語音指令進行分類。
4.  醫學診斷： 例如，根據醫學影像判斷疾病類型。
5.  金融風控： 例如，判斷交易是否為欺詐交易。
6.  推薦系統： 例如，根據用戶的歷史行為預測用戶可能感興趣的商品或內容。
7.  網路安全： 例如，檢測惡意軟體或網路攻擊。
**總結：**
分類深度神經網路是解決複雜分類問題的強大工具。它們能夠自動學習資料中的複雜模式，並在許多領域取得了令人矚目的成果。然而，訓練和部署這些模型需要大量的資料、計算資源和專業知識。在實際應用中，需要仔細考慮資料的特性、任務的要求以及可用的資源。

```python
# Classifiction DNN 分類深度神經網路
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

#DATA

```python
# Data
from tensorflow.keras.datasets import fashion_mnist
(x_train_set, y_train_set), (x_test, y_test) = fashion_mnist.load_data()
# Check Data
print(x_train_set.shape, y_train_set.shape)
print(x_test.shape, y_test.shape)
```

```python
#畫出某一個y_train的圖
i = 0
print(y_train_set[i])
plt.imshow(x_train_set[i], cmap='binary')
plt.show()
#
```

螢幕擷取畫面 2025-04-08 101445.png

```python
# Splite Data(分割資料)
from sklearn.model_selection import train_test_split
x_train, x_valid, y_train, y_valid = train_test_split(
    x_train_set, y_train_set, test_size=0.25, random_state=1)
```

```python
# Preprocessing(預處理)
x_train = x_train/255.0 #原本x值是0-255變成0-1
x_valid = x_valid/255.0
x_test = x_test/255.0
```

#Build Model

```python
# 1 Build Model 建構模型
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten

keras.backend.clear_session() #把先前做的一些設定全部把它clear做一個reset
np.random.seed(1)
tf.random.set_seed(1)
x_train.shape , x_valid.shape, x_test.shape
```

```python
model = Sequential([
    #Flatten(input_shape=x_train.shape[1:]), #把input_shape這個x_train它的這個第一個維度之後也就是28*28把它攤平所以28*28就會攤平成784這樣的一個維度
    keras.Input(shape=x_train.shape[1:]), # 明確定義輸入形狀
    Flatten(),# 現在 Flatten 不需要 input_shape ，Flatten層在深度學習中用來將多維數據攤平成一個一維向量
    Dense(300, activation='relu'),
    Dense(200, activation='relu'),
    Dense(100, activation='relu'),
    Dense(10, activation='softmax') #最後一個輸出層也是用Dense層，我們有10個類別所以就用10個神經元，輸出層統一用softmax這個觸發函數，把這10個類別的預測值用機率來表示
 ])
```

```python
# Check Model
model.summary()  # 顯示模型摘要
```

螢幕擷取畫面 2025-04-08 105457.png

#Compile

```python
# 2 Compile 在深度學習中，模型需要在訓練前進行編譯，以確定使用的優化器、損失函數，以及需要追蹤的評估指標。
model.compile(optimizer='sgd', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
```

optimizer='sgd': 使用梯度下降法（Stochastic Gradient Descent, SGD）作為模型的優化器。這是一種經典的優化方法，可以用來更新模型的權重以最小化損失函數。

loss='sparse_categorical_crossentropy': 使用稀疏分類交叉熵作為損失函數。這個損失函數適用於多分類問題，並且目標標籤是整數（非 one-hot 編碼）。例如，如果目標標籤是 0, 1, 2 而不是 [1, 0, 0], [0, 1, 0], 它可以避免使用 one-hot 編碼。

metrics=['accuracy']: 設定評估指標為「準確率」（accuracy），即模型在預測時正確的百分比。這能幫助追蹤模型在訓練和驗證過程中的表現。

#Train

```python
# Train
train = model.fit(x_train, y_train, epochs=20, validation_data=(x_valid, y_valid))
#batch_size=32(每一小批次訓練幾個樣本，預設是32個)
#epochs 進行幾個回合
#validation_data 驗證損失是否下降
```

```python
#將訓練後的資料畫成圖
pd.DataFrame(train.history).plot()
plt.grid(True)
plt.show()
#X軸:epochs 進行幾個回合
#Y軸:loss與accuracy 值
```

#Evaluate(評估)

```python
# 4 Evaluate(評估)
print(x_test.shape)
model.evaluate(x_test, y_test) #test資料有102筆，10000/32=312.多，所以分313批評估，平均accuracy: 0.8668
```

#Predict

```python
# 5 Predict
y_proba = model.predict(x_test)
y_proba[:10].round(2) #顯示前3筆各類別預測機率
```

```python
y_pred = np.argmax(y_proba, axis=1) #查找陣列中最大值的索引，axis=1：要求np.argmax在y_proba的每一列（axis=1）中查找最大值。
y_pred[:10] , y_test[:10]
```

```python
# confusion matrix
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```
